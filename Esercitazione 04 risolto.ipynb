```python
# Hint2: Potrebbe servirvi per verificare, senza usare un ciclo for, se gli elementi di un array assumono valori contenuti in una lista 
import numpy as np

mioArray = np.array([1,2,3,4,5,6,7])
lista = [2,5,9]
print(np.isin(mioArray,lista))
```

```python
# Solution

arraySex  =  dfNew['sex'].values
arraySmokeV = dfNew['smokes'].values
m = np.shape(arraySmokeV)[0]
arraySmoke = np.full(m,"",dtype = object)

for i in range(m):
    if arraySmokeV[i] == 'no':
        arraySmoke[i] = "NoSMOKER"
    else:
        arraySmoke[i] = "SMOKER"
    

groupSex = np.unique(arraySex)
groupSmoke = np.unique(arraySmoke)

matrice = np.zeros( (len(groupSex), len(groupSmoke)) ) # Creo tabella di contingenza

for sex in range(len(groupSex)):
    for smoke in range(len(groupSmoke)):
        matrice[sex,smoke] = np.sum( (arraySex == groupSex[sex]) & (arraySmoke == groupSmoke[smoke]) )

print('Tabella di contingenza')        
pd.DataFrame(matrice, index = groupSex, columns = groupSmoke)         
```

### 1.3

Rappresentate graficamente i risultati della tabell di contingenza tramite un bar plot

```python
# Solution
# rappresentazione tramite bar plot
trace1 = go.Bar(
    x=groupSex,
    y=matrice[:,0],
    name=groupSmoke[0]
)
trace2 = go.Bar(
    x=groupSex,
    y=matrice[:,1],
    name=groupSmoke[1]
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='group',
    xaxis=dict(title="Sesso"),
    yaxis=dict(title="numero di utenti")
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```

### 2.3 (opzionale)

L'istogramma generato al punto precedente che rappresenta la distribuzione delle età degli utenti disoccupati non è ben visibile nella stessa figura dell'istogramma che rappresenta la distribuzione delle età degli utenti lavoratori poichè il numero di utenti disoccupati è molto piu' basso rispetto ai lavoratori. Di conseguenza, anche le barre dell'istogramma sono molto piu' basse rispetto alle barre dell'istogramma caratterizzante gli utenti lavoratori. In questo caso, sarebbe meglio "normalizzare" gli istogrammi, dove l'altezza delle barre non è la frequenza assoluta (numero di utenti in un intervallo di età), ma la frequenta relativa (frazione di utenti in un intervallo di età). La cella sotto mostra come generare un istogramma normalizzato (basta impostare l'attributo ``histnorm`` uguale a ``probability'`` quando si creare l'oggetto della classe ``go.Histogram``).  

```python
#Hint
x = np.random.randn(1000)

trace = go.Histogram(x = x, histnorm = 'probability', marker = dict(color = 'rgb(255,0,0)'))
layout = go.Layout(title = 'Esempio di istogramma normalizzato', xaxis = dict(title = 'x'), yaxis = dict(title = 'frequenze relative'))
fig = go.Figure([trace], layout)
py.iplot(fig)
```

Generate adesso gli istogrammi normalizzati dell'età dei disoccupati e dei lavoratori.

```python
# Solution

hist1 = go.Histogram(x =  ageArray[jobArray == 'YES'],
              name = "Occupato",
             histnorm = 'probability',
             marker = dict(color = 'rgb(255,0,0)')         
             )

hist2 = go.Histogram(x =  ageArray[jobArray == 'NO'],
              name = "Disoccupato",
              histnorm='probability',
              marker = dict(color = 'rgb(0,0,255)')       
             ) 
 

layout = go.Layout(title = 'Histogramma normalizzato: Distribuzione delle età', xaxis = dict(title ='età'), yaxis = dict(title = 'Frequenze relative'))

groupBox = go.Figure(data = [hist1,  hist2], layout = layout)

py.iplot(groupBox)
```
