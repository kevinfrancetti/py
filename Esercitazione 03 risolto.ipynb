Torniamo adesso ai mondiali FIFA 2018 e carichiamo la [Tabella](https://www.fifa.com/worldcup/statistics/players/shots) con i tiri effettuati da ogni giocatore. Se proviamo ad eseguire i comandi della cella sopra, il server HTTP ci restituisce un errore di tipo ``HTTP Error 403: Forbidden`` (in parole povere, il server HTTP non ha accettato la richiesta di fare il parsing del file HTML). Senza entrare nel dettaglio, facciamo in modo che la nostra richiesta al server venga inviata come verrebbe inviata tramite un browser. Per fare questo basta eseguire i comandi della cella sotto. 

```python
import pandas as pd
import requests

url = 'https://www.fifa.com/worldcup/statistics/players/shots'

header = {
  "User-Agent": "Chrome/50.1.3672",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)

listaDf = pd.read_html(r.text)
df = listaDf[0]
```

### 1.1
Si generi un nuovo array (che chiameremo arrayShotsPerGame) che calcoli, per ogni giocatore, il numero di tiri a partita (dove la partita si intende $90$ minuti di gioco). 

Esempio 1: Il giocatore *Fuoriclasse* ha giocato durante il mondiale 135 minuti (una partita e mezzo) e ha effettuato 19 tiri, allora il numero di tiri a partita è uguale a
$19*\frac{90}{135} = 12.66$. 

Esempio 2: Il giocatore *Campione* ha giocato durante il mondiale solo 5 minuti e ha effettuato $4$ tiri, allora il numero di tiri a partita è uguale a
$4*\frac{90}{5} = 72$!!! (Durante il mondiale ha tirato solo $4$ volte, ma ha giocato molto poco!!!!). 

Hint: il numero di minuti giocati da ogni giocatore si trova nell colonna con label ``Minutes Played  MinP``.


```python
# Solution

arrayShotsPerGame = df['Shots  S']/df['Minutes Played  MinP']*90 # Questo non è ancora un numpy array
arrayShotsPerGame = arrayShotsPerGame.values
print(arrayShotsPerGame)
```

### 1.3

Plottare istogramma e boxplot delle osservazioni. Nominate opportunamente gli assi e se volete visualizzare meglio i grafici cambiate il range degli assi. Notate la presenza di outliers? Riuscite a spiegare perchè la media è piu' alta della mediana?

```python
# Solution
import plotly
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True) # All'inizio del vostro script scrivete questo comando per visualizzare i grafici sul notebook

# Genero istogramma
trace1 = go.Histogram(x = arrayShotsPerGame)
layout = go.Layout(title = 'Histogram',
                  xaxis = {'title': 'Tiri per partita'},
                  yaxis = {'title': 'No. di calciatori'} 
                  )
figure = go.Figure(data = [trace1], layout = layout)
plotly.offline.iplot(figure)

# Genere box-plot
trace2 = go.Box(x = arrayShotsPerGame)
layout = go.Layout(title = 'box-and-whisker plot',
                  xaxis = {'title': 'Tiri per partita'}
                  )
figure = go.Figure(data = [trace2], layout = layout)
plotly.offline.iplot(figure)
help(plotly.graph_objs._figure.Figure)
```

### 1.4
Calcolare lo z-score delle osservazioni e identificate gli outliers. Sapete spiegare la ragione di questi outliers?

```python
# Solution

zscore = (arrayShotsPerGame - np.mean(arrayShotsPerGame))/np.std(arrayShotsPerGame,ddof = 1)

zscoreAbs = np.abs(zscore) # Valore assoluto dello z-score

indiciOutliers = zscoreAbs >= 3

# Visualizziamo gli outliers 
print(f"Giocatori identificati come outliers:\n {arrayPlayers[indiciOutliers]} \n")

# Quanti minuti hanno giocato i giocatori identificati come otliers? E quanti tiri hanno effettuato?
arrayMinutes = df['Minutes Played  MinP'].values
arrayShots = df['Shots  S'].values

print(f"Minuti giocati dagli outliers:\n {arrayMinutes[indiciOutliers]} \n")
print(f"Tiri effettuati dagli outliers:\n {arrayShots[indiciOutliers]} \n")

nuovoArrayShotsPerGame = arrayShotsPerGame[zscoreAbs < 3]

print("Questi giocatori hanno giocato molto poco, ma in quei pochi minuti che hanno giocato "
      "hanno tirato relativamente tante volte. Il giocatore 'Arnor TRAUSTASON' ad esempio ha giocato 1 minuto e in quel minuto ha fatto un tiro. Quindi, in termini relativi, risulta aver"
       " effettuato 90 tiri a partita!  \n")

print("Questi giocatori non sono quindi rappresentativi delle mie osservazioni in quanto hanno giocato per troppo poco tempo")
```
